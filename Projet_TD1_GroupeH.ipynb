## Au blé d’or

On s’intéresse dans ce sujet à la gestion d’une boulangerie, qui veut planifier ses achats de matière première et la quantité de fabrication de ses baguettes et viennoiseries. Cette boulangerie réalise sur place des baguettes, des pains au chocolat et des croissants.

On se fournit en $m$ matières premières pour fabriquer $p$ produits. On note $r_i$ $(i = 1, \dots, m)$ la quantité commandée de la matière première $i$ et $q_j$ la quantité fabriquée du produit $j$. Pour produire une unité du produit $j$ $(j = 1, \dots, p)$, on suppose que l’on a besoin d’une quantité $a_{ij}$ de la matière première $i$. En définissant la matrice

$$
A = (a_{ij})_{1 \le i \le m, 1 \le j \le p} \in \mathcal{M}_{m,p}(\mathbb{R}_+),
$$

$$
r = (r_i) \in \mathbb{R}^m_+, \quad q = (q_j) \in \mathbb{R}^p_+,
$$

ceci se traduit par la relation

$$
r \ge A q
$$

Enfin, on note $d_j$ la demande pour le produit $j = 1, \dots, p$, $v_j$ son prix de vente et $c_i$ le coût unitaire d’achat de la matière première $i$. On cherche alors à maximiser les revenus de la boulangerie, qui s’expriment comme

$$
v^T \min(q, d) - c^T r
$$

où $v = (v_j) \in \mathbb{R}^p_+$, $d = (d_j) \in \mathbb{R}^p_+$, $c = (c_i) \in \mathbb{R}^m_+$ et le $\min$ s’entend comme le minimum composante par composante :

$$
\min(q, d) = (\min(q_i, d_i))_{1 \le i \le p}.
$$


## 1 Étude du problème d’optimisation

**Question 1**

 Interpréter le coût ci-dessus et en particulier le terme $\min(q, d)$.



**Question 1 Réponse**

Le profit de la boulangerie se définit comme les revenus moins les coûts d'où la définition de la fonction. $\min(q,d)$ signifie que pour chaque produit $j$, si on a une quantité supérieure à la demande, on peut fournir tous les clients et sinon, on écoule tout le produit. Ainsi, $v^T \min(q,d)$ représente les revenus totaux obtenus par les ventes et $c^T r$ représente le total des coûts d'achat de matière première.

**Question 2**

Quelle difficulté présente ce dernier terme dans le cadre d’un algorithme d’optimisation ?




**Question 2 Réponse**

La fonction $\min$ sur $\mathbb{R}^2$ que l'on peut définir par $\min(a,b)=\frac{a+b-|a-b|}{2}$ n'est pas différentiable sur $\mathbb{R}^2$ à cause de la valeur absolue, ce qui pose un problème de convergence des simulations (comme vu en cours par la méthode de descente de gradient par exemple). 


**Question 3**

En lieu et place de l'expression précédente, on se propose plutôt de maximiser la fonction

$$
v^T h(q, d) - c^T r
$$

où $h : \mathbb{R}^{2p} \to \mathbb{R}^p$ a pour $i$-ème composante

$$
h_i(q, d) = \frac{q_i e^{-\alpha q_i} + d_i e^{-\alpha d_i}}{e^{-\alpha q_i} + e^{-\alpha d_i}}
$$

avec $\alpha > 0$ une constante donnée. Justifier que, pour $\alpha \gg 1$, $h$ est une bonne approximation de la fonction min. Quel intérêt a-t-on à considérer ce problème approché plutôt que le problème originel ?




**Question 3 Réponse**

On prend $\alpha \gg 1$ et $i$ entre $1$ et $p$. 

- Si $0<q_i < d_i$, alors $e^{-\alpha q_i} \gg e^{-\alpha d_i}$ et donc $h_i(q,d) \approx q_i$
- Si $0 < d_i < q_i$, alors $e^{-\alpha d_i} \gg e^{-\alpha q_i}$ et donc $h_i(q,d) \approx d_i$
- Si $d_i=q_i$, on a directement et exactement $h_i(q,d)=q_i=d_i$.

Ainsi, la fonction $h$ représente bien une approximation de $\min(q,d)$ dans le cas $\alpha \gg 1$. L'introduction de cette approximation devient intéressant car la fonction $h$ est différentiable.

**Question 4**

On s’intéresse au problème approché. Formuler le problème d’optimisation à résoudre sous la forme

$$
\min_{c(z) \le 0} f(z)
$$

On précisera les variables de décision $z$, leur nombre $n$, les contraintes $c$ ainsi que la fonction objectif $f$ à minimiser.



**Question 4 Réponse**

Le problème initial revenait à maximiser la fonction profit ($v^T \min(q, d) - c^T r$) de notre boulangerie qui s'intéresse à la quantité de matières premières à acheter $r$ ainsi que la quantité à produire $q$. Ainsi la variable de décision est : $$z=(r,q) \in (\mathbb{R_+})^{m+p}$$ On suppose donc que les coûts d'achat de matière première $c$, ainsi que la matrice $A$, le prix de vente $v$ et la demande $d$ sont tous donnés et vérifient les contraintes de coefficients positifs. 

La fonction $h$ approxime la fonction $\min$, d'où en considérant le problème approché, on cherche bien à maximiser $v^T h(q, d) - c^T r$.

Ainsi, il revient donc à minimiser l'opposé de la quantité précédente, autrement dit : 
$$ f(z) = c^Tr - v^T h(q, d) $$

Les contraintes sont donc : $$c(z)=(-z,Aq-r)$$

## 2 Étude et résolution numérique

**Question 5**

Quelles méthodes de résolution peuvent être envisagées pour ce problème ?



**Question 5 Réponse**

Ici, la fonction $f$ est différentiable, donc on se trouve en présence d'un problème d'optimisation différentiable sous contrainte inégalité. On peut donc mettre en place les résolutions par :
- la méthode de KKT : on introduit des multiplicateurs de Lagrange pour caractériser les points optimaux
- la méthode du problème dual : on introduit un lagrangien et on peut étudier son problème dual en recherchant des points selles. 
- la méthode de descente de grandient : on met en place la méthode de descente de gradient en vérifiant la condition de contrainte inégalité à chaque pas. 

**Question 6**

On s’intéresse à $p = 3$ (baguettes de pain, pains au chocolat et croissants) produits et $m = 5$ ingrédients (levure, farine, sucre, beurre et chocolat). Développer un algorithme de résolution pour les valeurs numériques suivantes

$$
\alpha = 0.1,
$$

$$
c = 10^{-3} \begin{pmatrix} 30 \\ 1 \\ 1.3 \\ 4 \\ 1 \end{pmatrix}, \quad
v = \begin{pmatrix} 0.9 \\ 1.5 \\ 1.1 \end{pmatrix}, \quad
d = \begin{pmatrix} 400 \\ 67 \\ 33 \end{pmatrix},
$$

$$
A =
\begin{pmatrix}
3.5 & 2 & 1 \\
250 & 80 & 25 \\
0 & 8 & 3 \\
0 & 40 & 10 \\
0 & 8.5 & 0
\end{pmatrix}
$$

Tester différentes valeurs de $\alpha$. Les résultats obtenus sont-ils conformes à votre intuition ?



**Question 6 Réponse**

On reprend la méthode de résolution pour un problème d'optimisation sous contrainte d'après le TP 3 avec la syntaxe de casadi pour un problème non linéaire (en raison de la fonction $h$).

```python
import casadi as ca
import numpy as np

#Données de l'énoncé
c=np.array([30, 1, 1.3, 4, 1])*1e-3
v=np.array([0.9, 1.5, 1.1])
d=np.array([400, 67, 33])
A=np.array([[3.5, 2, 1],[250, 80, 25],[0, 8, 3],[0, 40, 10],[0, 8.5, 0]])
m,p=5,3 

#Variables sur lesquelles on travaille
def solution(alpha):
    r=ca.MX.sym("r", m)
    q=ca.MX.sym("q", p)

    #Fonction minimum approchée
    def min(qi, di):
        return (qi*ca.exp(-alpha*qi)+di*ca.exp(-alpha*di))/(ca.exp(-alpha*qi)+ca.exp(-alpha*di))
    h=ca.vertcat(*[min(q[i], d[i]) for i in range(p)])

    #Quantité à minimiser 
    f=ca.dot(c, r)-ca.dot(v, h)
    #Contraintes
    g=ca.vertcat(r,q,r-A@q)
    lbg=np.zeros(g.shape[0])
    ubg=np.inf*np.ones(g.shape[0])

    #Solveur
    nlp={"x": ca.vertcat(r,q), "f": f, "g": g}
    solver=ca.nlpsol("solver", "ipopt", nlp)


    #On résout
    x0=np.concatenate([A@d,d])
    sol=solver(x0=x0, lbg=lbg, ubg=ubg)

    x_opt = sol["x"].full().flatten()
    r_opt=x_opt[:m]
    q_opt=x_opt[m:]
    return r_opt,q_opt
R,Q=[],[]
for alpha in [0.1,0.5,0.75,0.87] :
    r_opt,q_opt = solution(alpha)
    R.append(r_opt)
    Q.append(q_opt)
print(R,Q)
```

**Conclusions**

On s'aperçoit que plus $\alpha$ augmente, plus on se rapproche de $q_{opt}=d$ par valeurs supérieures. Ceci est cohérent avec nos suppositions, plus $\alpha$ augmente, plus la fonction $h$ approxime bien la fonction $\min$, et de fait, on se rapproche de l'optimum recherché. ## 3 Étude du problème non-régularisé

On considère toujours que $K$ scenarii sont possibles, mais correspondant cette fois à la fonction coût
non-approchée initiale.

**Question 8 (a)**  Formuler le problème d’optimisation correspondant.

**Question 8 (a) Réponse**

On remplace simplement $h$ par la vraie fonction $\min$. On doit donc résoudre en $z=(r,q)$ 

$$\min_{c(z)\leq 0 } (f(z)) = \min_{\substack{r \geq Aq \\ r \geq 0 \\ q \geq 0}} \left( c^Tr-\sum_{k=1}^{K} \pi_k (v^T \min(q, d^k)) \right)

**Question 8 (b)** Montrer que la contrainte $r \ge Aq$ est active à l’optimum.

**Question 8 (b) Réponse**

On remarque que $$f(z)=c^Tr-\sum_{k=1}^{K} \pi_k (v^T \min(q, d^k))$$ possède deux termes indépendants. 

En effet : 
- $c^Tr$ ne dépend que de $r$ donc ne dépend pas de $q$
- $\sum_{k=1}^{K} \pi_k (v^T \min(q, d^k))$ ne dépend que de $q$ donc ne dépend pas de $r$

Ainsi, puisque toutes les quantités sont positives, pour minimiser la fonction $f$, on désire rendre $c^Tr$ le plus petit possible et $\sum_{k=1}^{K} \pi_k (v^T \min(q, d^k))$ le plus grand possible. Ces quantités étant indépendantes, rendre $c^Tr$ le plus petit possible revient donc à rendre $r$ le plus petit possible sous la contrainte $r \geq Aq$, donc : $$r=Aq$$

Ainsi la contrainte est bien active à l'optimum.

**Question 8 (c)** En déduire que l’on peut éliminer la variable $r$ du problème d’optimisation et formuler un problème d’optimisation sans contrainte.


**Question 8 (c) Réponse**

D'après la réponse précédente, à l'optimum, la contrainte $r \geq Aq$ est active, i.e. $r=Aq$. Ainsi, en remplaçant $r$ par $Aq$ on obtient un problème d'optimisation sans contrainte avec une seule variable à savoir $q$. On peut donc bien éliminer la variable $r$ du problème d'optimisation. 

Le nouveau problème devient : 

$$ \min_{\substack{r \geq Aq \\ r \geq 0 \\ q \geq 0}} \left( c^Tr-\sum_{k=1}^{K} \pi_k (v^T \min(q, d^k)) \right) = \min_{q \geq 0 } \left( c^TAq-\sum_{k=1}^{K} \pi_k (v^T \min(q, d^k)) \right)$$

Ainsi, on obtient un problème de minimisation sur la seule variable $q$, sur son domaine de définition, à savoir $q \in \mathbb{R}^p$.

**Question 9 (a)** Montrer que maximiser $v^T \min(q, d^k)$ est équivalent à maximiser $v^T u^k$ sous les contraintes $u^k \le d^k$ et $u^k \le q$.

**Question 9 (a) Réponse**

Pour chaque scénario $1\leq k \leq K$, on souhaite montrer que :
$$\max_{q \geq 0 } \left( v^T \min(q, d^k) \right)=\max_{\substack{q \geq 0 \\ u_k \geq 0 \\ u_k \leq d^k \\ u_k \leq q}} (v^T u_k) $$

On veut donc montrer que pour $q$ fixé : $$v^T \min(q, d^k) = \max_{\substack{u_k \geq 0 \\ u_k \leq d^k \\ u_k \leq q}} (v^T u_k)$$

On a à $q$ fixé : $$\max_{\substack{u_k \geq 0 \\ u_k \leq d^k \\ u_k \leq q}} (v^T u_k) = \max_{\substack{u_k \geq 0 \\ u_k \leq min(d^k,q)}} (v^T u_k) = v^T \min(q, d^k) $$
car $v\geq 0$.

**Question 9 (b)** Reformuler le problème d’optimisation de la Question $8$ à l’aide de ces nouvelles variables de décision. Quel intérêt présente cette nouvelle formulation ?

**Question 9 (b) Réponse**

D'après la question 8, le problème devient en le réécrivant, puisque toutes les quantités sont positives, en notant $d=(d^k)_{1 \leq k \leq K}$ et $u=(u_k)_{1 \leq k \leq K}$, on a  :

$$ 
\min_{q \geq 0 } \left( c^TAq-\sum_{k=1}^{K} \pi_k (v^T \min(q, d^k)) \right) = \min_{\substack{q \geq 0 \\ u \geq 0 \\ u \leq d \\ u \leq q}}\left(c^TAq - \sum_{k=1}^{K} \pi_k (v^T u_k) \right) 
$$ 
L'intérêt de cette nouvelle formulation est que l'on remplace un problème d'optimisation non lisse par la présence d'un $\min$ et sans contrainte, sauf celle portant sur $q$, par un problème d'optimisation avec contrainte linéaire en introduisant de nouvelles variables $(u_k)_{1 \leq k \leq K}$.

**Question 9 (c)** Résoudre ce problème pour les valeurs numériques fournies en Question $7$. Comparer avec les résultats précédents.

**Question 9 (c) Code**

```python
import casadi as ca
import numpy as np

# Données de l'énoncé
c=np.array([30,1,1.3,4,1])*1e-3
v=np.array([0.9,1.5,1.1])
d=np.array([[400,67,33], [500,80,53], [300,60,43]])
PI=np.array([0.5,0.3, 0.2])
A=np.array([[3.5,2,1], [250, 80,25], [0,8,3], [0,40,10], [0,8.5,0]])
m,p,K=5,3,3 

def solution_ter():
    #variables
    q=ca.MX.sym('q', p)
    u=[ca.MX.sym(f'u_{k}', p) for k in range(K)]
    variables=ca.vertcat(q, *u)
    
    #fonction f objectif
    sum=0
    for k in range(K):
        sum+=PI[k]*ca.dot(v, u[k])
    f=ca.dot(c, A @ q)-sum
    
    #Contraintes
    g=[]
    for k in range(K):
        g.append(q-u[k])  
        g.append(d[k]-u[k])
    g = ca.vertcat(*g)
    
    #bornes sur variables
    lbx=np.zeros(variables.shape[0])
    ubx=np.inf*np.ones(variables.shape[0])
    
    #bornes sur g
    lbg= np.zeros(g.shape[0])
    ubg= np.inf*np.ones(g.shape[0])
    
    
    lp= {'x': variables, 'f': f, 'g': g}
    solver= ca.qpsol('solver', 'highs', lp)
    
    #Résolution
    x0=np.zeros(variables.shape[0]) 
    x0[:p]=d[0]#q = premier scénario et u_k nuls en premier lieu
    sol=solver(x0=x0, lbx=lbx, ubx=ubx, lbg=lbg, ubg=ubg)
    x_opt=sol['x'].full().flatten()
    q_opt=x_opt[:p]
    r_opt=A@q_opt #contrainte active à l'optimum
    return r_opt, q_opt

r_opt,q_opt=solution_ter()
print(f"q optimal : {q_opt}")
print(f"r optimal : {r_opt}")
```

**Conclusions**

On obtient bien ce qu'on attendait à la question 7 lorsqu'on augmentait $\alpha$ à savoir que les quantités à produire suivent les scénarios dominants : le premier pour la baguette, et le deuxième pour le croissant et le pain au chocolat. 

Le programme converge bien vers la valeur optimale.

**Question 10 (b) Code**

```python
import casadi as ca
import numpy as np


# Données de l'énoncé
c=np.array([30,1,1.3,4,1])*1e-3
v=np.array([0.9,1.5,1.1])
d=np.array([[400,67,33], [500,80,53], [300,60,43]])
PI=np.array([0.5,0.3, 0.2])
A=np.array([[3.5,2,1], [250, 80,25], [0,8,3], [0,40,10], [0,8.5,0]])
m,p,K=5,3,3 

def solution_NL(x0):
    #Variables
    q = ca.MX.sym('q', p)
    #Fonction f objectif (non lisse)
    Sum=0
    for k in range(K):
        h_k=ca.vertcat(*[ca.fmin(q[i], d[k][i]) for i in range(p)])
        Sum += PI[k]*ca.dot(v, h_k)
    f = ca.dot(c,A@q)-Sum
    #Contraintes
    g=q 
    lbg=np.zeros(p)
    ubg=np.inf*np.ones(p)
    
    nlp = {'x': q, 'f': f, 'g': g}    
    solver= ca.nlpsol('solver', 'ipopt', nlp)
    sol=solver(x0=x0, lbg=lbg, ubg=ubg)
    q_opt=sol['x'].full().flatten()
    return q_opt
Q=[]
for x0 in [np.zeros(3),d[0],d[1],d[2],sum(PI[k]*d[k] for k in range(K))]:
    q_opt= solution_NL(x0)
    Q.append(q_opt)
print(Q)
```

**Résultats Code Propres**

- Pour $x_0=[0,0,0]$ : $q_{opt}=[400., 80.13 , 53.10]$

- Pour $x_0 = d^1$ : $q_{opt}=[400., 118.63, 53.44]$

- Pour $x_0 = d^2$ : $q_{opt}=[400.,  93.44, 54.41]$

- Pour $x_0 = d^3$ : $q_{opt}= [400., 129.53, 53.00]$

- Pour $x_0 = \sum_{k=1}^K (\pi_k d^k)$ : $q_{opt}= [400., 88.63, 101.55]$

**Conclusions**

On remarque donc que suivant le point de départ $x_0$ les valeurs à convergence de l'optimum diffèrent en raison des sous-gradient qui amènent à des plateaux et des stagnations de résolution alors même que nous ne sommes pas arrivés à la solution optimale des questions précédentes. Il n'est donc pas préférable, quand on résout un problème d'optimisation avec des fonctions objectifs non-lisses, de s'attaquer directement à la résolution du problème avec des sous-gradients qui ne renvoie pas forcémement l'optimum désiré. Cette résolution a donc mis en exergue les limites pratiques des solveurs dans un cas simple. 

Dans ce projet, nous avons en premier lieu approximer la fonction non-lisse par une fonction lisse. Cette méthode :
- approxime certes le problème initial
- ne dépend pas du point de départ
- renvoie un optimum approché (en raison de la valeur de $\alpha$ et de l'approximation de $\min$ par la fonction $h$) mais il reste fidèle

Enfin, pour atteindre une valeur d'optimum exacte, nous avons vu dans la question $9$ une méthode de linéarisation garantissant la convergence vers l'optimum. En transformant le problème non-lisse en introduisant de nouvelles variables $(u_k)_{1\le k \le K}$, on obtenait une résolution exacte du problème. L'avantage de cette méthode est d'ajouter des variables pour rendre la résolution plus abordable (de  non-lisse non-linéaire à lisse linéaire).
